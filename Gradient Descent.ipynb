### Gradient Descent


```python
% matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
data_path = "Data\\pga.csv"
pga = pd.read_csv(os.path.relpath(data_path))

"""
Because distance and accuracy have different scales, we need to
normalize distance and accuracy to avoid measurements of Euclidean distance
being biased towards the larger value
"""

pga.distance = (pga.distance - pga.distance.mean()) / pga.distance.std()
pga.accuracy = (pga.accuracy - pga.accuracy.mean()) / pga.accuracy.std()

pga.head()
```

```python
plt.scatter(pga.distance, pga.accuracy)
plt.xlabel('normalized distance')
plt.ylabel('normalized accuracy')
plt.show()
```

##### Looks like we have a negative slope on this bad boy

So, we can use scikit-learn's LinearRegression class to fit a model to our normalized data
The model will be:

$accuracy_i=\theta_i distance_i + \theta_0 + \epsilon_i$

where:
$\theta_i$ and $\theta_0$ are coefficients and $\epsilon_i$ is the error term

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

# add new axis for matrix operations
distance = pga.distance[:,np.newaxis]

# fit the model
lr.fit(distance, pga.accuracy)

# assign feature coefficient to theta1
theta1 = lr.coef_[0]

print(theta1)
```

### But what about big dataaaaaa

The sklearn LinearRegression class uses OLS by default, which finds coefficients deterministically. However, this can be expensive if we have a lot of features and rows. 

Gradient Descent is used in cases where we aren't able to fit our data in memory. 

Gradient descent uses cost functions to minimize residuals iteratively by randomly generating coefficients and then updating the coefficients to reduce the error term for each coefficient.

For some model $h_\theta = \theta_1 x+ \theta_0$,

The cost function is defined as:

$J(\theta_0 , \theta_1 )=\frac{1}{2m} \sum_{i=1} ^m (h_\theta(x_i )-y_i )^2$

Which just means the cost function is half the average difference between the predicted and observed value. Gradient descent will randomly choose coefficients and then update them to minimize the cost.


```python

def cost(theta0 ,theta1, x, y):
        # Initialize cost
    J = 0
    # The number of observations
    m = len(x)
    # Loop through each observation
    for i in range(m):
        # Compute the hypothesis 
        h = theta1 * x[i] + theta0
        # Add to cost
        J += (h - y[i])**2
    # Average and normalize cost
    J /= (2*m)
    return J

theta0 = 100

# create random array of coefficients
theta1s = np.linspace(-3,2,100)

# calculate costs for each of theta1s coefficients

costs = [cost(theta0, i, pga.distance, pga.accuracy) for i in theta1s]

plt.plot(theta1s, costs)
plt.xlabel('theta1s coefficients')
plt.ylabel('costs')

plt.show()

```

### Looks like we got ourselves a parabola

So when running through the theta1s coefficients, we have can see that the result is a quadratic function with a global minimum

Therefore, the goal is to find a set of parameters that produces the global minimum, which is the point where the model has the lowest error

Because we are finding a minimum of error, $\theta_1$ and $\theta_0$, we can visualize the descent on a 3-dimensional surface where the x and y axis represent the slope of the model and the z-axis will be the cost.

We can use the matplotlib 3d api's [Axes3D](https://matplotlib.org/mpl_toolkits/mplot3d/api.html) to visualize the surface

```python
from mpl_toolkits.mplot3d import Axes3D

# we'll need to represent each possible pair of the x and y values
# numpy's meshgrid will produce an n-dimensional coordinate matrices to represent the pairings
# create theta0s values
theta0s = np.linspace(-2,2,100)
theta1s = np.linspace(-2,2,100)
Xtheta0s, Ytheta1s = np.meshgrid(theta0s, theta1s)

COST = np.empty(shape=(100,100))

for i in range(100):
    for j in range(100):
        COST[i,j] = cost(Xtheta0s[0,i], Ytheta1s[j,0],pga.distance,pga.accuracy)

```

```python
fig = plt.figure(figsize = (8,8))

ax = fig.gca(projection='3d')
ax.plot_surface(X=Xtheta0s,Y=Ytheta1s,Z=COST)
plt.show()
```

### The the $\partial$readed $\partial$

So now we have this beautiful gradient surface and we want to find the global minimum. 
Because it's a surface, we need to find the minimum slope of the surface with respect to Z (cost) in terms of each of the parameters, $\theta_0$ and $\theta_1$

So we want to find both:
In terms of $\theta_0$:

$\dfrac{\partial J(\theta_0, \theta_1)}{\partial \theta_0} = \dfrac{1}{m} \sum_{i=1}^m (h_{\theta}(x_i)-y_i)$ 

and in terms of $\theta_1$ 

$\dfrac{\partial J(\theta_1 )}{\partial \theta_1}=\dfrac{1}{m} \sum_{i=1}^m (h_{\theta}(x_i)-y_i)*x_i$

Where $h_\theta$ represents the model's hypothesized prediction $h_\theta = \theta_1 x + \theta_0$



```python
# Let's definte partial cost functions for theta0 and theta1 respectively

def partial_cost_theta1(theta0, theta1, x, y):
    # hypothesis (prediction)
    h = theta1*x + theta0
    # hypothesis minus observed * x
    diff = (h-y)*x
    
    # compute partial by averaging
    partial = diff.sum() / (x.shape[0])
    return partial

def partial_cost_theta0(theta0, theta1, x, y):
    # hypothesis (prediction)
    h = theta1*x + theta0
    # hypothesis minus observed
    diff = (h-y)
    
    # compute partial by averaging
    partial = diff.sum() / (x.shape[0])
    return partial

partial_theta1 = partial_cost_theta1(1, 1, pga.distance, pga.accuracy)

partial_theta0 = partial_cost_theta0(1, 1, pga.distance, pga.accuracy)


print("Partial theta0: {}".format(partial_theta0))
print("Partial theta1: {}".format(partial_theta1))


    
```

### Snooping around the gradient

Now that we have functions that define the partial cost of the coefficients, we can define an algorithm that:

- randomly generates coefficients
- calculates partial derivative of each
- moves in the direction of the steepest slope 
    - this essentially causes the gradient descent algorithm to follow the slope that leads to a local (hopefully global) minimum
    
For two variables, the algorithm looks like this:

$$Repeat\:until\:convergence: \{\\
\theta_1 := \theta_1 - \alpha * \dfrac{\partial J(\theta_0, \theta_1)}{\partial \theta_0} \\
\theta_0 := \theta_0 - \alpha * \dfrac{\partial J(\theta_0, \theta_1)}{\partial \theta_1} \\
\\ \}$$

Stepping through the algorithm:
We are familiar with $\theta_0$ and $\theta_1$ 

which in the case of gradient descent algoriths stand for the current values assigned to the coefficients in the model.

$\alpha$ is the learning rate, which is $\Re[0,1]$. It essentially means the size of the step that the algorithm takes in the direction of the steepest slope. 

Assigning $\alpha$ is tricky, since if it is too small, it will coverge really slowly and we might not arrive anywhere near the global minimum, if it is too large, it risks bouncing around above and below the optimal values for $\theta_0$ and $\theta_1$

So the algorithm generates coefficients, subtracts the cost function times the learning rate ($\alpha$) from the coefficient to produce the next set of coefficients. 



```python
# let's make dis algorithm

def gradient_descent(x, y, alpha=0.1, theta0 = 0, theta1 = 0):
    # epochs are the number of iterations to run through the algorithm
    max_epochs = 1000
    counter = 0
    # intial cost
    c = cost(theta0, theta1, x, y)
    costs = [c]
    
    '''
    define convergence threshold
    when the difference between the previous cost and the current cost
    is less than the convergence threshold, we consider the
    parameters to be converged
    '''
    
    convergence_threshold = 0.000001
    
    cprev = c + 10
    theta0s = [theta0]
    theta1s = [theta1]
    
    while (np.abs(cprev - c) > convergence_threshold) and (counter < max_epochs):
        cprev = c
        # update with partial costs
        update0 = alpha * partial_cost_theta0(theta0, theta1, x, y)
        update1 = alpha * partial_cost_theta1(theta0, theta1, x, y)
        
        theta0 -= update0
        theta1 -= update1
        
        theta0s.append(theta0)
        theta1s.append(theta1)
        
        c = cost(theta0, theta1, x, y)
        
        costs.append(c)
        counter += 1
    return dict(theta0 = theta0, theta1 = theta1, costs = costs)


    
```

```python
descent_results = gradient_descent(pga.distance, pga.accuracy, alpha = 0.01)

plt.plot(range(len(descent_results['costs'])), descent_results['costs'])
plt.xlabel('iterations')
plt.ylabel('costs')

plt.show()

```
