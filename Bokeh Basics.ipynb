```python
# Plot an iris histogram
from ipywidgets import interact
from bokeh.io import push_notebook, output_notebook, show 
from bokeh.charts import Histogram
from bokeh.sampledata.iris import flowers as data
from bokeh.plotting import figure
output_notebook()
```

```python
x=[1,2,3,4,5]
y=[6,7,2,4,5]
# output to static HTML file

p = figure(title="sample line",x_axis_label='x',y_axis_label='y')
p.line(x,y, legend='Temp', line_width=2)
#show in browser notebook
show(p)
```

```python
# Create histogram
hist = Histogram(data, values='petal_length', color='species',legend='top_right',bins=12)
show(hist)
```

```python
sc = figure(plot_width=400, plot_height=400)
# create scatter with circles
# all bokeh markers can take size in terms of screen size units as an argument
sc.circle(list(range(1,6)), [6,7,2,4,5], radius=.2, line_color='navy', fill_color='orange',fill_alpha=0.5)
show(sc)
```

```python
import numpy as np
N = 20 
img = np.empty((N,N), dtype=np.uint32)
view = img.view(dtype=np.uint8).reshape((N,N,4))
for i in range(N):
    for j in range(N):
        view[i, j, 0] = int(i/N*255) # red
        view[i, j, 1] = 158          # green
        view[i, j, 2] = int(j/N*255) # blue
        view[i, j, 3] = 255          # alpha

# Create a new plot (with a fixed range) using figure
p = figure(x_range=[0,10], y_range=[0,10])

# add an RGBA image renderer
p.image_rgba(image = [img],x=[0],y=[0],dw=[10],dh=[10])

show(p)

        
```

```python
# calling more than one glyph on a figure
# import ColumnDataSource class
from bokeh.models import ColumnDataSource

# Create ColumnDataSource object from pandas DataFrame
source = ColumnDataSource(data)

df_fig = figure()
df_fig.circle('petal_length','petal_width',source=source)
show(df_fig)
data.head()
# General observation; I don't know what the ColumnDataSource is for
```

```python
species = list(set(data['species'].values))
print(species)
virginica = data[data['species']=='virginica']
setosa = data[data['species'] == 'setosa']
versicolor = data[data['species']=='versicolor']
flowers = figure()
flowers.circle(x=virginica['petal_length'], y=virginica['petal_width'], size=5, alpha=0.4, line_color='red',fill_color=None, line_width=2)
flowers.triangle(x=setosa['petal_length'], y=setosa['petal_width'], size=5, alpha=0.4, line_color='blue',fill_color=None, line_width=2)
flowers.square(x=versicolor['petal_length'], y=versicolor['petal_width'], size=5, alpha=0.4, line_color='green', line_width=2)
show(flowers)
```
